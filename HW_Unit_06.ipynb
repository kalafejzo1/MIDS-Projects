````python
# TODO: Adjust the path below as necessary. This assumes the file is in the same directory as this script.

SOWPODS_PATH = "sowpods.txt"

def read_sowpods(fpath: str) -> list[str]:
    """Read the SOWPODS dictionary file and return a set of words."""
    with open(fpath, "r") as f:
        return [line.strip("\n") for line in f.readlines()]


print(f"Read {len(read_sowpods(SOWPODS_PATH))} words from SOWPODS file.")
````

````python
# Q6-2 Grading Tag:

scores = { "a": 1, "c": 3, "b": 3, "e": 1,  "d": 2, "g": 2,
                 "f": 4, "i": 1, "h": 4, "k": 5,  "j": 8, "m": 3,
                 "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
                 "r": 1, "u": 1, "t": 1, "w": 4,  "v": 4, "y": 4,
                 "x": 8, "z": 10 }
def score_word(word: str) -> int:
    """Calculates and returns the Scrabble score for that word using the provided `scores` dictionary. """
    # TODO: Implement score_word here.
    sum = 0
    lower_case = word.lower()
    for char in lower_case:  #Loops through every character in the word
        if (char in scores.keys()):  #Checks if character is a letter
            sum += scores[char]  #Adds the current character score to the total
    return sum   


def validate_rack(rack: str) -> bool:
    """Returns `True` if the rack is valid according to the rules (2-7 characters, only A-Z, a-z, or `*`), and `False` otherwise."""
    # TODO: Implement validate_rack here.
    if (len(rack)>7 or len(rack)<2):  #Returns False if the length of the rack is less than 2 or greater than 7
        return False
    lower_case = rack.lower()
    for char in lower_case:  #Returns False if the characters in the rack are not valid
        if not (char in scores.keys() or char == "*"):
            return False
    return True
    


def run_scrabble(rack: str, words: list[str]) -> dict[str, int]:
    # TODO: Implement run_scrabble here.
    lower_rack = rack.lower()  #Uppercases rack so that all letters will match the uppercases in words
    dic = {}
    for letter in lower_rack:  #Creates a dictionary of all of the letters in rack and how many times they appear
        if (letter in dic.keys()):
            dic[letter] += 1
        else:
            dic[letter] = 1
    def valid_word(word):  
        """Returns a dictionary with a boolean as the key and a string as the value. The boolean is set to True if the word can be made with the 
        rack and False otherwise. The value is an empty string if no * replacements were made for letters, but if * replacements were made, then
        the string value is the letters that were replaced."""
        temp_dic = dic.copy()
        return_dic = {}
        sub_let = ""
        lower_word = word.lower()
        for letter in lower_word:
            if letter not in temp_dic and "*" not in temp_dic:  #Checks if the letter isn't in upper rack and if * isn't in the rack
                return {False: ""}
            elif letter not in temp_dic and "*" in temp_dic.keys():  #Checks if the letter isn't in upper rack but rack has * available
                if (temp_dic["*"] == 1):  #Removes the * from the dictionary if it's used
                    temp_dic.pop("*")
                    sub_let += letter
                else:  #Decreases the number of the * available by 1
                    temp_dic["*"] -= 1
                    sub_let += letter
            else:  #If the letter is in the dictionary, removes it/decreases the letter from the dictionary
                    if(temp_dic[letter] == 1): #Removes the letter from the dictionary
                        temp_dic.pop(letter)
                    else:  #Decreases the number of the letter available by 1
                        temp_dic[letter] -= 1
        return {True: sub_let}
    word_dic = {}
    for word in words:
        if (list(valid_word(word).keys()))[0] == True:  #Checks if the word is a valid word
            if (list(valid_word(word).values())[0] != ""):  #Case for words with replacements
                word_dic[word] = score_word(word) - score_word(list(valid_word(word).values())[0])
            else:  #Case for no replacements
                word_dic[word] = score_word(word)
    return word_dic
    


if __name__ == "__main__":
    # The condition above checks if this script is being run directly or being imported. More info:
    # https://docs.python.org/3/library/__main__.html
    # https://realpython.com/if-name-main-python/#in-short-it-allows-you-to-execute-code-when-the-file-runs-as-a-script-but-not-when-its-imported-as-a-module
    words = read_sowpods(SOWPODS_PATH)
    rack = input()  # feel free to use input() to get the rack from the user.

    # TODO: Call the functions above to complete the scrabble solver
    if (validate_rack(rack) == True):
        run_scrabble(rack, words)
````
