````python
# QP1-1 Grading Tag:

import random

class LocalView:
    """This class represents the area around an animal that the animal is able to see. It stores the directions that the animal is able to move in or the option to stay'"""
    def __init__(self, can_move_north = False, can_move_east = False, can_move_south = False, can_move_west = False):
        self.move_directions = ["stay"]
        if can_move_north == True:
            self.move_directions.append("north")
        if can_move_south == True:
            self.move_directions.append("south")
        if can_move_east == True:
            self.move_directions.append("east")
        if can_move_west == True:
            self.move_directions.append("west")
        
    def __repr__(self):
        string = " ".join(str(direction) for direction in self.move_directions)
        return string
    
    def __str__(self):
        string = " ".join(str(direction) for direction in self.move_directions)
        return string


class Animal:
    """Creates the super class Animal that takes in a LocalView object"""
    def __init__(self, icon):
        if len(icon)!=1:
            raise Exception("icon must be a single character.")
        elif type(icon) is not str:
            raise Exception("icon must be a single character string.")
        else:
            self.icon = icon
        
    def __repr__(self):
        return self.icon

    def __str__(self):
        return self.icon


    def move(self, moves):
        """Takes in a LocalView and picks a move based off of the possible options"""
        pass

class Zebra(Animal):
    """Creates the subclass Zebra"""
    def __init__(self, icon = "\U0001F993"):
        self.icon = icon
        super().__init__(self.icon)

    def move(self, moves):
        """Takes in a LocalView and randomly picks a move based off of the possible options"""
        return random.choice(moves.move_directions)

class Monkey(Animal):
    """Creates the subclass Monkey"""
    def __init__(self, icon = "\U0001F412"):
        self.icon = icon
        super().__init__(self.icon)

    def move(self, moves):
        """Takes in a LocalView and prioritizes North, then East, then South, then West, then stay"""
        if 'north' in moves.move_directions:
            return 'north'
        elif 'east' in moves.move_directions:
            return 'east'
        elif 'south' in moves.move_directions:
            return 'south'
        elif 'west' in moves.move_directions:
            return 'west'
        else:
            return 'stay'
        
class Zoo:

    def __init__(self, height, width):
        self.height = height
        self.width = width
        self.zoo_matrix = [[Animal("x") for i in range(self.width)] for j in range(self.height)] #Creates a zoo matrix of empty Animals with width and height]

    def add_animal(self, animal, height_cord, width_cord):
        """Adds an animal to the zoo at a given height and width"""
        print(self.zoo_matrix)
        self.zoo_matrix[height_cord][width_cord] = animal
        print (self.zoo_matrix)

    def time_step(self):
        """Moves each animal one at a time as long as they don't move out of the grid or move into a spot occupied by another `Animal`."""
        i = 0
        moves = []
        while i < self.height:
            j = 0
            while j < self.width: #Creates the list of possible moves for each animal
                north = False
                south = False
                west = False
                east = False
                if self.zoo_matrix[i][j].icon != "x":
                    if i != 0 and self.zoo_matrix[i-1][j].icon == "x": #Checks for north
                        north = True
                    if i != self.height - 1 and self.zoo_matrix[i+1][j].icon == "x": #Checks for south
                        south = True
                    if j != 0 and self.zoo_matrix[i][j-1].icon == "x": #Checks for west
                        west = True
                    if j != self.width - 1 and self.zoo_matrix[i][j+1].icon == "x": #Checks for east
                        east = True
                    local_view = LocalView(north, east, south, west)
                    move = self.zoo_matrix[i][j].move(local_view) #Selects the move for each animal
                    #Reindexes the animals so they move in the zoo
                    if str(move) == "north":
                        indecies = [i, j, i-1, j] #Stores the old i, old j, new i and new j as a list
                        moves.append(indecies)
                    elif str(move) == "south":
                        indecies = [i, j, i+1, j] #Stores the old i, old j, new i and new j as a list
                        moves.append(indecies)
                    elif str(move) == "west":
                        indecies = [i, j, i, j-1] #Stores the old i, old j, new i and new j as a list
                        moves.append(indecies)
                    elif str(move) == "east":
                        indecies = [i, j, i, j+1] #Stores the old i, old j, new i and new j as a list
                        moves.append(indecies)
                j+=1
            i+=1
        for indecies in moves:
            self.zoo_matrix[indecies[2]][indecies[3]] = self.zoo_matrix[indecies[0]][indecies[1]]
            self.zoo_matrix[indecies[0]][indecies[1]] = Animal("x")
        
    def __repr__(self):
        string = ""
        for j in range(self.height):
            for i in range(self.width):
                string += " " +str(self.zoo_matrix[j][i].icon)
            string += "\n"
        return string

    def __str__(self):
        string = ""
        for j in range(self.height):
            for i in range(self.width):
                string += " " +str(self.zoo_matrix[j][i].icon)
            string += "\n"
        return string 
    
    def position_of_animal(self, animal):
        """Finds the coordinates (position) of the animal passed in as an attribute"""
        i = 0
        while i < self.height:
            j = 0
            while j < self.width: 
                if self.zoo_matrix[i][j] == animal:
                    return (i, j)
                j+=1
            i+=1
        raise Exception(f'{animal.icon} not in zoo')
    
    def animal_at_position (self, height_cord, width_cord):
        """Takes in the height and width coordinate of an animal and returns True if contains and animal and none otherwise"""
        if self.zoo_matrix[height_cord][width_cord].icon != "x":
            return True
                 
# Zoo display using tkinter
import tkinter as tk

class ZooDisplay(tk.Tk):
    def __init__(self, zoo):
        super().__init__()
        self._zoo = zoo
        self._text_widget = tk.Text(self)
        self._text_widget.configure(font=("Times", 20, "bold"))
        self._text_widget.pack()
        self._text_widget.insert(tk.END, "Replace with Text you wish to display.")
        self._text_widget.after(1000, self.update)

    def update(self):
        self._zoo.time_step()
        self._text_widget.delete("1.0", "end")
        self._text_widget.insert(tk.END, str(self._zoo))
        self._text_widget.after(1000, self.update)
    
````
